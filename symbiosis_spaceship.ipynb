# Types of sampling:
1. Random smapling:\
Probability is also known as random sampling
Why random sampling is also known as probability sampling?
  Each element in the population has an equal chance of being selected or known probability of being selected(Stratified random sampling)

2. Stratified random sampling:\
the population is devided into groups of strata. A random sample is selected from each stratum based upon the percentage that each subgroup represents population. Stratified random samples are generally more accurate in representing the population than are simple random samples.

3. Convenience sampling:\
 Convenience samples are nonprobability samples. Therefore it is not possible to specify the probability of any population element's being selected for the sample. Indeed, it is not possible to specify the population from which the sample was drawn.
